# Local Interaction

This notebook demonstrates how to study local interaction model using the **[`localint`](https://github.com/oyamad/game_theory_models/tree/localint)** Python library.

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from __future__ import division
from localint import LocalInteraction

%matplotlib inline
```

## 1. Parameters

The **`localint`** class requires two parameters, **payoff matrix** and **adjacency matrix**.

### Payoff Matrix

Payoff matrix must be 2-dimensional square numpy array.   In a game-theoretic model, it means that both the set of actions and the payoff function are the same across all players.

For instance, consider a coordination game where the payoff table is given by the following:

1$\backslash$2 | $A$           | $B$ 
 ------------- |---------------| ---------
 $A$           | 4, 4          |    0, 2   
 $B$           | 2, 0          |   3, 3       

Note that this payoff table represents that the game is symmetric. Because of the symmetricity, it suffices to record the only one of the player's payoffs like the following:

```python
payoff_matrix = np.asarray([[4, 0], [2, 3]])
print payoff_matrix
```

### Adjacency Matrix

Adjacency matrix represents how the nodes connect. Particularly, in the context of the local interaction model, it represents whether each pair of players interacts and how strong the interaction of each pair of players is.

Let's consider an adjacency matrix given by the following:
\begin{equation}
[a_{ij}] = 
\begin{bmatrix}
0 &1 &3\\
2 &0 &1\\
3 &2 &0
\end{bmatrix}
\end{equation}

For example, $a_{12}(=1)$ denotes the weight of player 2's action on player 1. Note that the weight of player 1's action on player 2 ($a_{21}=2$) is different. That is, the **`localint`** class allow adjacency matrix to be asymmetric unlike payoff matrix.

## 2. Exapmles (under construction)

In the following, I give you some examples of typical graphs, and analyze the local interaction models corresponding to those graphs.

```python
def drawg(li, pos=None):
    G = nx.Graph(li.adj_matrix)
    if pos is None:
        pos=nx.circular_layout(G)
    nx.draw_networkx_nodes(G, pos, nodelist=G.nodes(), node_size=500, node_color='w')
    nx.draw_networkx_edges(G, pos, arrows=True)

    labels={}
    for i, action in zip(G.nodes(), li.current_actions):
        labels[i] = r'${0}$'.format(action)
    nx.draw_networkx_labels(G, pos, labels)

    plt.axis('off')
    plt.show()
```

### Ex-1. Complete graph

```python
G1 = nx.complete_graph(8)
nx.draw_circular(G1, node_color='w', with_labels=False)
```

```python
li1 = LocalInteraction(payoff_matrix, nx.adjacency_matrix(G1))
```

```python
init_actions = np.zeros(8, dtype=int)
init_actions[range(3)] = 1
print init_actions
```

```python
li1.set_init_actions(init_actions)
print li1.current_actions
```

```python
for i in range(3):
    drawg(li1)
    li1.play()
```

```python
for i in range(500):
    np.random.seed(i)
    if all(li1.simulate(ts_length=6, init_actions=[1,1,1,0,0,0,0,0], revision='sequential')[-1] == 1):
        break
print i
```

```python
np.random.seed(32)
li1.simulate(ts_length=6, init_actions=[1,1,1,0,0,0,0,0], revision='sequential')
```

### Ex-2. Path graph

```python
G2 = nx.path_graph(8)
pos = {}
for i in range(8):
    pos[i] = np.array([i/7, 0])
```

```python
li2 = LocalInteraction(payoff_matrix, nx.adjacency_matrix(G2))
```

```python
drawg(li2, pos)
```

```python
init_actions = np.zeros(8, dtype=int)
init_actions[[0,7]] = 1
print init_actions
```

```python
li2.set_init_actions(init_actions)
print li2.current_actions
```

```python
drawg(li2, pos)
while any(li2.current_actions[i] != 1 for i in range(8)): 
    li2.play()
    drawg(li2, pos)
```

```python
for i in range(2000):
    np.random.seed(i)
    if all(li2.simulate(ts_length=20, init_actions=[1,0,0,0,0,0,0,0], revision='sequential')[-1] == 1):
        break
print i
```

```python
np.random.seed(169)
li2.simulate(ts_length=20, init_actions=[1,0,0,0,0,0,0,0], revision='sequential')
```

### Ex-3. Star graph

```python
G3 = nx.ladder_graph(4)
pos = {0: np.array([0, 1]), 1: np.array([0.25, 1]), 2: np.array([0.5, 1]), 3: np.array([0.75, 1]),
       4: np.array([0, 0]), 5: np.array([0.25, 0]), 6: np.array([0.5, 0]), 7: np.array([0.75, 0])}
```

```python
li3 = LocalInteraction(payoff_matrix, nx.adjacency_matrix(G3))
```

```python
drawg(li3, pos)
```

```python
init_actions = np.zeros(8, dtype=int)
init_actions[[0,1,2,6]] = 1
print init_actions
```

```python
li3.set_init_actions(init_actions)
print li3.current_actions
```

```python
for i in range(3):
    drawg(li3, pos)
    li3.play()
```

```python
for i in range(1000):
    np.random.seed(i)
    if all(li3.simulate(ts_length=7, init_actions=[0,1,0,0,0,0,1,0], revision='sequential')[-1] == 1):
        break
print i
```
